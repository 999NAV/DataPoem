```python
# Extract year, month, and week
df2['Year'] = df2['Date/Time'].dt.year
df2['Month'] = df2['Date/Time'].dt.month
df2['Week'] = df2['Date/Time'].dt.isocalendar().week

# Group by year, month, and week and calculate the mean for each group
monthly_data = df2.groupby(['Year', 'Month'])[['Max Temp (°C)', 'Min Temp (°C)', 'Mean Temp (°C)']].mean()
weekly_data = df2.groupby(['Year', 'Week'])[['Max Temp (°C)', 'Min Temp (°C)', 'Mean Temp (°C)']].mean()

# Plot Max Temp, Min Temp, and Mean Temp with respect to year
monthly_data.plot(y=['Max Temp (°C)', 'Min Temp (°C)', 'Mean Temp (°C)'], subplots=True, title='Yearly Temperature Trends')
plt.show()

# Plot Max Temp, Min Temp, and Mean Temp with respect to month
monthly_data.plot(y=['Max Temp (°C)', 'Min Temp (°C)', 'Mean Temp (°C)'], subplots=True, title='Monthly Temperature Trends')
plt.show()

# Plot Max Temp, Min Temp, and Mean Temp with respect to week
weekly_data.plot(y=['Max Temp (°C)', 'Min Temp (°C)', 'Mean Temp (°C)'], subplots=True, title='Weekly Temperature Trends')
plt.show()
```

```python
# Group by year, month, and week and calculate the mean for each group
monthly_data = df2.groupby(['Year', 'Month'])[['Total Snow (cm)', 'Total Precip (mm)']].mean()
weekly_data = df2.groupby(['Year', 'Week'])[['Total Snow (cm)', 'Total Precip (mm)']].mean()

# Plot Total Snow (cm) and Total Precip (mm) with respect to year
monthly_data.plot(y=['Total Snow (cm)', 'Total Precip (mm)'], subplots=True, title='Yearly Snow and Precipitation Trends')
plt.show()

# Plot Total Snow (cm) and Total Precip (mm) with respect to month
monthly_data.plot(y=['Total Snow (cm)', 'Total Precip (mm)'], subplots=True, title='Monthly Snow and Precipitation Trends')
plt.show()

# Plot Total Snow (cm) and Total Precip (mm) with respect to week
weekly_data.plot(y=['Total Snow (cm)', 'Total Precip (mm)'], subplots=True, title='Weekly Snow and Precipitation Trends')
plt.show()
```

```python
# Assuming you have a 'Date' column in your DataFrame
df2['Date/Time'] = pd.to_datetime(df2['Date/Time'])

# Set 'Date' column as the index
df2.set_index('Date/Time', inplace=True)

# Now, resample and calculate the mean
df_final = df2.resample('D').mean()
df_final.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_final)

# Define sequence length and features
sequence_length = 10  # Number of time steps in each sequence
num_features = len(df_final.columns)

# Create sequences and corresponding labels
sequences = []
labels = []
for i in range(len(scaled_data) - sequence_length):
    seq = scaled_data[i:i+sequence_length]
    label = scaled_data[i+sequence_length][6]  # '_tempm' column index
    sequences.append(seq)
    labels.append(label)

# Convert to numpy arrays
sequences = np.array(sequences)
labels = np.array(labels)

# Split into train and test sets
train_size = int(0.8 * len(sequences))
train_x, test_x = sequences[:train_size], sequences[train_size:]
train_y, test_y = labels[:train_size], labels[train_size:]

print("Train X shape:", train_x.shape)
print("Train Y shape:", train_y.shape)
print("Test X shape:", test_x.shape)
print("Test Y shape:", test_y.shape)
```

#FNN Model

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Define the FNN model
model = Sequential()
model.add(Flatten(input_shape=(sequence_length, num_features)))  # Flatten the input sequences
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))  # Linear activation for regression

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
batch_size = 16
epochs = 50
model.fit(train_x, train_y, validation_data=(test_x, test_y), batch_size=batch_size, epochs=epochs)

# Evaluate the model
test_loss = model.evaluate(test_x, test_y)
print(f"Test Loss: {test_loss}")

# Make predictions
predictions = model.predict(test_x)
```

```python
# Plot the actual and predicted values
plt.figure(figsize=(8, 6))
plt.plot(test_y, label='Actual', marker='o')
plt.plot(predictions, label='Predicted', marker='x')
plt.title('Actual vs. Predicted Values')
plt.xlabel('Data Point')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

## Boosting Methods

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Create and train the Gradient Boosting Regressor
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
model.fit(train_x.reshape(train_x.shape[0], -1), train_y)  # Reshape train_x to 2D

# Make predictions
predictions = model.predict(test_x.reshape(test_x.shape[0], -1))  # Reshape test_x to 2D

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(test_y, predictions)
print(f"Mean Squared Error: {mse}")

# Plot the actual vs. predicted values
plt.figure(figsize=(10, 6))
plt.plot(test_y, label='Actual', marker='o')
plt.plot(predictions, label='Predicted', marker='x')
plt.title('Actual vs. Predicted Values (Gradient Boosting)')
plt.xlabel('Data Point')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a mask to identify zero values in test_y
zero_mask = test_y == 0

# Calculate the absolute percentage error for each data point
absolute_percentage_errors = np.abs((test_y - predictions) / np.where(zero_mask, 1, test_y))

# Calculate the Mean Absolute Percentage Error (MAPE)
mape = np.mean(absolute_percentage_errors[~zero_mask]) * 100  # Exclude zero values

print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```
